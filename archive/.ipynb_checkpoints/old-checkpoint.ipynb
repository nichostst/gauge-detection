```python
import cv2
import glob
import matplotlib.pyplot as plt
import numpy as np
import os


# Settings
dirpath = './images'
dim = (256, 256)
```

```python
def rename_imgs(dirpath):
    imgs = glob.glob(dirpath + '/*')
    for i, img in enumerate(imgs):
        fmt = img.split('.')[-1]
        os.rename(img, f"{dirpath}/{i}.{fmt}")


def get_grayscale_diff(gray):
    resized = cv2.resize(gray, dim)

    foo = np.empty([x+2 for x in resized.shape])
    foo[:] = np.nan

    f = {}
    f['up'] = foo.copy()
    f['down'] = foo.copy()
    f['left'] = foo.copy()
    f['right'] = foo.copy()

    foo[1:-1, 1:-1] = resized
    f['up'][2:, 1:-1] = resized
    f['down'][:-2, 1:-1] = resized
    f['left'][1:-1, 2:] = resized
    f['right'][1:-1, :-2] = resized

    # Get 4-way mean absolute difference of the pixel value
    diff = np.nanmean(np.array([np.abs(foo-x)[1:-1, 1:-1] for x in f.values()]), axis=0)
    return diff
```

```python
img = glob.glob(dirpath + '/*')[0]
raw = cv2.imread(img)
gray = cv2.cvtColor(raw, cv2.COLOR_RGB2GRAY)
```

```python
canny = cv2.Canny(gray, 40, 80)
plt.imshow(canny)
```

```python
diff = get_grayscale_diff(gray)
diff[diff < np.quantile(diff, 0.9)] = 0
plt.imshow(diff)
```

```python
cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 150, param1=80, param2=30, minRadius=60, maxRadius=256)
```
